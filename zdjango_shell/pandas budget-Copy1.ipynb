```python
import pandas as pd
```

```python
data = Lancamento.objects\
        .values('lancamento__title','lancamento__category__category', 'amount', 'amount2', 'amount3', 'amount4', 'amount5', 'amount6', 'amount7', 'amount8', 'amount9', 'amount10', 'amount11', 'amount12')\
        .order_by('lancamento__category__category')
```

```python
df = pd.DataFrame(data)
df
```

```python
df['total'] = df['amount'] + df['amount2'] + df['amount3'] \
+ df['amount4']+ df['amount5']+ df['amount6']\
+ df['amount7']+ df['amount8']+ df['amount9']\
+ df['amount10']+ df['amount11']+df['amount12']
df
```

```python
df['%'] = round(100 * df['amount'] / df['amount'].sum())
df['% 2'] = round(100 * df['amount2'] / df['amount2'].sum())
df['% 3'] = round(100 * df['amount3'] / df['amount3'].sum())
df['% 4'] = round(100 * df['amount4'] / df['amount4'].sum())
df['% 5'] = round(100 * df['amount5'] / df['amount5'].sum())
df['% 6'] = round(100 * df['amount6'] / df['amount6'].sum())
df['% 7'] = round(100 * df['amount7'] / df['amount7'].sum())
df['% 8'] = round(100 * df['amount8'] / df['amount8'].sum())
df['% 9'] = round(100 * df['amount9'] / df['amount9'].sum())
df['% 10'] = round(100 * df['amount10'] / df['amount10'].sum())
df['% 11'] = round(100 * df['amount11'] / df['amount11'].sum())
df['% 12'] = round(100 * df['amount12'] / df['amount12'].sum())
df
```

```python
df = df.rename(columns={'amount': 'mes1', 'amount2': 'mes2', 'amount3': 'mes3', 'amount4': 'mes4', 'amount5': 'mes5', 'amount6': 'mes6', 'amount7': 'mes7', 'amount8': 'mes8', 'amount9': 'mes9', 'amount10': 'mes10', 'amount11': 'mes11', 'amount12': 'mes12'})
df = df.rename(columns={'lancamento__category__category':'categoria'})
```

```python
pivot_table = pd.pivot_table(df, values=['mes1','%','mes2','% 2', 'mes3', '% 3', 'mes4', '% 4','mes5', '% 5', 'mes6', '% 6', 'mes7', '% 7', 'mes8', '% 8', 'mes9', '% 9', 'mes10', '% 10', 'mes11', '% 11', 'mes12', '% 12'], index=['categoria'], fill_value=0, aggfunc=sum)
pivot_table
```

```python
cols = list(pivot_table.columns.values)
cols
```

```python
newdf = pivot_table[[
'mes1', '%',
'mes2', '% 2',
'mes3', '% 3',
'mes4', '% 4',
'mes5', '% 5',
'mes6', '% 6',
'mes7', '% 7',
'mes8', '% 8',
'mes9', '% 9',
'mes10', '% 10',
'mes11', '% 11',
'mes12', '% 12',
]]
newdf.loc['TOTAL DESPESAS']= newdf.sum()
```
